```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_tavily import TavilySearch
from google.colab import userdata


LLM=ChatGoogleGenerativeAI(model="gemini-2.5-flash",temperature=0.7,google_api_key=userdata.get('project2'))
tavily_search = TavilySearch(result="general",max_results=5,tavily_api_key=userdata.get('tavily'))
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

parser=StrOutputParser()
Extract_Prompt=PromptTemplate(
    template="""You are a precise web content extraction assistant.

Your task is to process each URL provided and extract its main textual content as a separate document.

Instructions:
- Treat each URL independently.
- Extract only meaningful content (ignore ads, navigation menus, footers, scripts).
- Preserve important headings and structure.
- Do NOT summarize.
- Do NOT merge documents.
- Do NOT add explanations.
- If content cannot be retrieved, return "Content not accessible".

Output Format:
Return a Python-style list where each item corresponds to one URL:

[
  "<clean extracted text>",...
]

URLs:
{urls}""",input_varabile=["urls"])

summary_prompt=PromptTemplate(
    template="""You are a precise summarization assistant.

Your task is to summarize each document separately.

Instructions:
- Treat each document independently.
- Produce a clear, concise summary.
- Do NOT merge documents.
- Do NOT add extra commentary.
- Preserve key facts, numbers, names, and conclusions.
- If document is empty, return "No content available".

Output Format:
In bullet points
Return a Python list:

[
    "summary1","summary3",...
]

Documents:
{documents}""",input_variable=["documents"])

report_prompt=PromptTemplate(
    template="""You are an expert report writer.

Your task is to analyze multiple document summaries and produce one unified, well-structured report.

Instructions:
- Combine insights from all summaries.
- Identify common themes, differences, and key findings.
- Do NOT repeat the same information.
- Do NOT mention individual document IDs.
- Write in formal report style.
- Maintain logical flow.

Report Structure:
1. Title
2. Executive Summary
3. Key Findings
4. Comparative Insights
5. Important Statistics / Facts
6. Final Conclusion

Rules:
- Be concise but informative
- Avoid filler language
- Use clear headings
- Base report ONLY on provided summaries

Summaries:
{summaries}""",input_variable=["summaries"])

question="How can retrieval-augmented generation (RAG) architectures be optimized to reduce hallucinations in large language models while maintaining response latency suitable for real-time applications?"



chains=Extract_Prompt |LLM | parser | summary_prompt |LLM |parser|report_prompt|LLM|parser

```

```python
def Application(query):
  search_result=tavily_search.invoke(query)

  urls=[search_result["results"][i]["url"]  for i in range(0,5)]
  final_result=chains.invoke({"urls":urls})
  return final_result

user_input=input("Enter your research question:")
result=Application(user_input)
print(result)
```
